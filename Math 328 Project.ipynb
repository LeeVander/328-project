```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
Housing = pd.read_csv("AB_NYC_2019.csv")
Housing
```

```python
Housing.columns = ['id','name','host','host_name','neighborhood_group','neighborhood', 'latitude', 'longitude', 'room','price', "min_night","reviews", "last_review", "Per_month", "host_list", "availability"]
```

```python
len(Housing)
```

```python
Housing.describe()
#There is missing data in Per_month and last_review 
```

```python
Housing['neighborhood_group']
```

```python
Housing['neighborhood_group'].value_counts()
```

```python
House = Housing['neighborhood_group'].value_counts()
House
House.plot(kind ='bar')
plt.title('Neighborhoods')
plt.xlabel('Bouroughs')
plt.ylabel('per stay')
plt.show()
```

```python
prices = Housing['price']
prices
prices.plot()
plt.title('Price')
plt.xlabel('cost')
plt.ylabel('per registered stay')
plt.show()
```

```python
Ready_home = Housing['availability']
Ready_home
Ready_home.hist(bins = 20)
plt.title('Availability')
plt.xlabel('per home')
plt.ylabel('Airbnb')
plt.show()
```

```python
Stay = Housing['min_night']
Stay.hist(bins = 40)
plt.title('min Amount of Stays at Airbnb')
plt.xlabel('Amount of Stays')
plt.ylabel('per person')
plt.show()
```

```python
sns.scatterplot(x = "availability", y="price" ,data= Housing)
```

```python
Housing.plot.scatter(x = "price", y = "availability")
plt.title("price vs availability")
plt.xlabel("price")
plt.ylabel("availability")
plt.show()
```
